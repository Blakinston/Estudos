Mlib Api: spark + machine learing

```
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression()
```

```
from pyspark.ml.feature import VectorAssembler
```

```
d.columns
```

```
assm = VectorAssembler(
                    inputCols=['Income','Limit','Rating','Cards'],
                    outputCol="saida")
assm# sepracao 
```

```
d2 = assm.transform(d)
```

```
d3 = d2.select("Income","saida")
```

```
d3.show()
```

```
d4 = d3.withColumnRenamed("Income","label")
d4 = d4.withColumnRenamed("saida","features")
```

```
train,test = d4.randomSplit([0.8,0.2],seed=100)
```

```
mod = lr.fit(train)
```

```
pd = mod.transform(test)
```
