```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X =  iris.data
y =  iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)
```

```python
# criterion: gini ou entropia 
# splitter: random(mais)  
arvore = DecisionTreeClassifier(criterion='gini',max_depth=1)
arvore.fit(X_train,y_train)
```

```python
# atributo mais importante e o meio
arvore.feature_importances_
```

```python
import matplotlib.pyplot as plt
from sklearn import tree 

previsores = iris.feature_names
plt.figure(figsize=(20,20))
#filled, coloca cor na arvore
tree.plot_tree(arvore,feature_names=previsores,class_names=iris.target_names,filled=True);
```

```python
# Impureza de Gini é melhor do que a entropia para selecionar as melhores características.
# Entropia consegue balancear melhor a arvore
from sklearn.metrics import accuracy_score

pred = arvore.predict(X_test)
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred));
# na classe 1, ele indentifica 100% das flores que pertencem a classe e ele esta acertando 50% 
# 50% precisao e 100% recall
# F1-Score: média harmônica entre precisão e recall.grandezas inversamente proporcionais
```
