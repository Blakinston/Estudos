## Algoritimo apriori

Descobrir os conjuntos de itens com suporte, proporcao q aparece, definimos o minimo especifico<br>
descobrir regras apartir da confianca 

```python
from mlxtend.frequent_patterns import apriori,association_rules
ass = apriori(d2,min_support=0.05,use_colnames=True)
ass.sort_values("support",ascending=False).head(10)
```

```python
# antecedente: comprou pao, tambem cafe
r = association_rules(ass,metric="lift")
r
```

```python
# produtos que sao vendidos no minimo 4x ao dia
n_dias= 7 # numero de dias
n_compras=4 # vezes que um produto e comprado por dia
total_prod = 720

suporte = (n_compras * n_dias)/total_prod
suporte
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
ap = apriori(df,min_support=suporte,use_colnames=True) 
ap
```

```python
reg = association_rules(ap,metric="lift")
reg
# valiar pelo lift e confianca
```
