```python
def normal_inv(med,dvp,pct):
    '''desonhecemos o nosso N, e queremos achar o percentil  da nossa distribuicao
       probit: funcao responsavel por encontrar os valores da nossa distribuicoa baseado na media e dvp
       e percentil'''
    pct = stats.norm.ppf(pct)# achando o valor da porcentagem na tabela z
    v = dvp * pct 
    return round(v+med,0)
```

```python
# media de nivel de colesterol e de 210mg/dl 
# normamente distribuidos em um dvp 38.6mg/dl 
# qual o maior nivel de colesterol dentre os 1% dos homens nesse nivel 
normal_inv(210,38.6,0.01)
```

```python
normal_inv(1000,300,0.95) - normal_inv(1000,300,0.9999)  
stats.norm.ic
```

```python
envisados = skewnorm.rvs(a=-10,size=1000) # enviesamento negativo
sn.displot(envisados)
```

```python
alpha = 0.05

_,p=stats.shapiro(envisados)

if p>alpha:
    print('Distribuicao Normal')
else :
    print('Distribuicao nao normal')
```
