formula de Sturges: defini o numero de cortes no tamanho da basse<br>
K= 1 + 3.3 * logN<br><br>
k = classes ou numero de cortes <br> 
amplitude total = diferenca entre o maior e menor valor no rol<br>
intervalo =  e o valor no qual assumo para pegar de um ponto a outro [n,n+8)
                                                                    

```python
def sturges(dataset):
    ''' Formula de Sturges e amplitude geral e de classes
        K: numero de classes
        amp_class: intervalo entre limite inferior e superior
        amp: amplitude entre maximo e minimo 
    '''
    n = len(dataset)
    K = 1+3.3*np.log(n) # numero de classes,formula de sturges
    K = np.round(K).astype('int64')
    amp = np.max(dataset) - np.min(dataset) # amplitude geral
    amp_class = amp/K
    return K,amp_class,amp 
    
```

```python
k,_,_=sturges(alturas)
pd.cut(alturas,bins=k)
```

```python
amp = np.max(alturas) - np.min(alturas)
'Amplitude Geral: '+str(amp)
```

```python

'Amplitude de um intervalo de classe: '+str(amp/sturges(alturas))
```
