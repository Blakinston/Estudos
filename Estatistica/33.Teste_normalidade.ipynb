```python
import numpy as np 
from scipy.stats import norm 
v = norm.rvs(size=100000)
v[:5]
```

```python
from scipy.stats import shapiro 

# ano tem acurracia boa para amostra > 5000
_,p=shapiro(v[:4900])
if p<=0.05: 
    print('h0 rejeitada')
else:
    print('Ha rejeitada')
```

```python
import seaborn as sn 
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
sn.histplot(v[:4900],bins=30,kde=True)# distribuicao quase uniforme
```
