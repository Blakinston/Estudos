$f(x) = 2^x $

```python
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as  plt 
import math 

x = np.array([-2,-1,0,1,2,3]).astype('float64')
y = np.array([ math.pow(2,i) for i in x ]).round(1)
plt.plot(x,y)
```

log<sub>2</sub><sup>x</sup><br>

```python
# ou seja trocamos valor de x com de y  da funcao exponencial 
x2 = np.array([ math.log(i,2) for i in y ]).round(1) #o log do meu y da funcao exponencial resulta no meu x da funcao logaritimica
plt.plot(y,x2)
```

```python
# trocando valores de y,x temos uma funcao logaritimica
plt.plot(y,x)
```
