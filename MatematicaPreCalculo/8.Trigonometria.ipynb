Trigonometria: geometria dos triangulos

triangulo retangulo: tem um angulo de 90 graus<br>
hipotenusa: maior lado do triangulo, oposto ao angulo reto<br>
catetos: lados opostos a hipotenusa

Angulo reto: 90 graus<br>
alpha e beta: angulos agudos

teorema de tales diz que teremos 180 graus, dentro do triangulo

Seno: cateto oposto/Hipotenusa<br>
Coseno:cateto adjacente/Hipotenusa <br>
Tangente: cateto oposto/cateto adjacente

```python
import numpy as np 
import matplotlib.pyplot as plt 


plt.plot([0,0,1,0], [0,1,0,0])
```

```python
# seno 
# o que ela faz e tracar o plano baseadonos angulos passados

x = np.linspace(-2*np.pi,2*np.pi,100) # multiplicacao para se alcancar os radianos do meu circulo trignometrico,6.28
ys = np.sin(x)
plt.figure(figsize=(10,7))
plt.plot(x,ys)
```

```python
# coseno
yc = np.cos(x)
plt.figure(figsize=(10,7))
plt.plot(x,yc)
```

```python
# encontramos angulos como esses pontos extremos que sao assintoticos 
# sen(x) / cos(x)
yt = np.tan(x)
plt.figure(figsize=(10,7))
plt.plot(x,yt)
```
