```python
import matplotlib.pyplot as plt
import numpy as np
from functools import reduce 

v =  np.array([4,5])
v2 = np.array([-2,5])
soma = v+v2
```

```python
plt.figure(figsize=(9,7))
o = np.array([[0]])# origem
plt.quiver(*o,v[0],v[1],scale=45)
```

```python
merge = np.array([v,v2,soma])

plt.figure(figsize=(9,7))

o = np.array([[0,0,0],[0,0,0]])# origem

plt.quiver(*o,merge[:,0],merge[:,1],color=['r','b','g'],scale=45)
plt.show()
```
